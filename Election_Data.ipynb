```python
import csv
import numpy as np
import pandas as pd

def open_csv(filename, d = ','):
    data = []
    
    with open(filename, encoding = 'uft-8') as mData:
        info = csv.reader(mData, delimiter = d)
        
        for row in info:
            data.append(row)
        return data
    
def open_exel(filename):
    voter_cols = ['Precinct','BM','BF','BU','WM','WF','WU','APacM','APacF','APacU','HM','HF','HU','NM','NF','NU','OM','OF','OU','UNKNOWN MALE','UNKNOWN FEMALE','UNKNOWN','TOTAL VOTERS']
    column_range = "B,C,F,I,L,O,R,U,X,AA,AD,AG,AJ,AM, AP,AS, AV, AY, BB, BE,BH, BK,BN"
    return pd.read_excel(filename, header=None, names=voter_cols, index_col=0,  usecols=column_range, skiprows=3)

def population(df):
    df['POW_w_Un'] = df[['BM','BF','BU','APacM','APacF','APacU','HM','HF','HU','NM','NF','NU','OM','OF','OU','UNKNOWN MALE','UNKNOWN FEMALE','UNKNOWN']].sum()
    df['POW_wo_Un'] = df[['BM','BF','BU','APacM','APacF','APacU','HM','HF','HU','NM','NF','NU','OM','OF','OU']].sum()
    df['White'] = df[['WM','WF','WU']].sum()
    
```

```python
applingGA = open_exel('./files/GA_November8_2016(CountyPrecinct)/APPLING.xlsx')
population(applingGA)
# print(applingGA.columns)
print(applingGA)
sum_row = applingGA[['BM','BF','BU','WM','WF','WU','APacM','APacF','APacU','HM','HF','HU','NM','NF','NU','OM','OF','OU','UNKNOWN MALE','UNKNOWN FEMALE','UNKNOWN','TOTAL VOTERS']].sum()
df_sum=pd.DataFrame(data=sum_row).T
df_sum=df_sum.reindex(columns=applingGA.columns)
df_sum
# df_final=applingGA.append(df_sum,ignore_index=True)
# applingGA
# df_final

```
